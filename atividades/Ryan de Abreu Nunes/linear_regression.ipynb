```python
#=============================================================================
# Import packages
#=============================================================================
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Change the plotting style
plt.style.use('seaborn-v0_8-whitegrid')
```

# Warm-up exercise

```python
def warmUpExercise():
    """
    Create a 5x5 identity matrix.
    
    Returns:
    --------
    array-like
        5x5 identity matrix
    """
    return np.eye(5)
```

# Data visualization

```python
def plotData(X, y):
    """
    Plot the training data.
    
    Parameters:
    -----------
    X : array-like
        Training data features
    y : array-like
        Training data labels
    """
    plt.figure(figsize=(10, 6))
    plt.plot(X, y, 'rx', markersize=10, label='Training data')
    plt.xlabel('Population of City in 10,000s')
    plt.ylabel('Profit in $10,000s')
    plt.title('Training Data')
    plt.legend()
    plt.grid(True)
    plt.show()
```

# Cost function and gradient descent

```python
def computeCost(X, y, theta):
    """
    Compute cost for linear regression.
    
    Parameters:
    -----------
    X : array-like
        Training data features
    y : array-like
        Training data labels
    theta : array-like
        Model parameters
        
    Returns:
    --------
    float
        Computed cost
    """
    m = len(y)
    predictions = X.dot(theta)
    sqErrors = (predictions - y) ** 2
    J = 1/(2*m) * np.sum(sqErrors)
    return J
```

```python
def gradientDescent(X, y, theta, alpha, iterations):
    """
    Perform gradient descent to learn theta.
    
    Parameters:
    -----------
    X : array-like
        Training data features
    y : array-like
        Training data labels
    theta : array-like
        Initial model parameters
    alpha : float
        Learning rate
    iterations : int
        Number of iterations
        
    Returns:
    --------
    tuple
        Optimized theta and cost history
    """
    m = len(y)
    J_history = np.zeros(iterations)
    
    for i in range(iterations):
        predictions = X.dot(theta)
        errors = predictions - y
        theta = theta - (alpha/m) * X.T.dot(errors)
        J_history[i] = computeCost(X, y, theta)
    
    return theta, J_history
```

# Visualization functions

```python
def plotLinearFit(X, y, theta):
    """
    Plot the linear fit.
    
    Parameters:
    -----------
    X : array-like
        Training data features
    y : array-like
        Training data labels
    theta : array-like
        Model parameters
    """
    plt.figure(figsize=(10, 6))
    plt.plot(X[:, 1], y, 'rx', markersize=10, label='Training data')
    plt.plot(X[:, 1], X.dot(theta), 'b-', label='Linear regression')
    plt.xlabel('Population of City in 10,000s')
    plt.ylabel('Profit in $10,000s')
    plt.title('Linear Regression Fit')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
def plotCostHistory(J_history, iterations):
    """
    Plot the cost history.
    
    Parameters:
    -----------
    J_history : array-like
        Cost history
    iterations : int
        Number of iterations
    """
    plt.figure(figsize=(10, 6))
    plt.plot(range(iterations), J_history, 'b-')
    plt.xlabel('Iterations')
    plt.ylabel('Cost J')
    plt.title('Cost History')
    plt.grid(True)
    plt.show()
```

```python
def plotCostSurface(X, y):
    """
    Plot the cost function surface.
    
    Parameters:
    -----------
    X : array-like
        Training data features
    y : array-like
        Training data labels
    """
    # Grid over which we will calculate J
    theta0_vals = np.linspace(-10, 10, 100)
    theta1_vals = np.linspace(-1, 4, 100)
    
    # Initialize J_vals to a matrix of 0's
    J_vals = np.zeros((len(theta0_vals), len(theta1_vals)))
    
    # Fill out J_vals
    for i, theta0 in enumerate(theta0_vals):
        for j, theta1 in enumerate(theta1_vals):
            t = np.array([theta0, theta1])
            J_vals[i, j] = computeCost(X, y, t)
    
    # Surface plot
    fig = plt.figure(figsize=(12, 5))
    
    # Surface plot
    ax = fig.add_subplot(121, projection='3d')
    theta0_mesh, theta1_mesh = np.meshgrid(theta0_vals, theta1_vals)
    ax.plot_surface(theta0_mesh, theta1_mesh, J_vals.T, cmap='viridis')
    ax.set_xlabel('θ₀')
    ax.set_ylabel('θ₁')
    ax.set_zlabel('J(θ₀, θ₁)')
    ax.set_title('Cost Function Surface')
    
    # Contour plot
    ax = fig.add_subplot(122)
    ax.contour(theta0_vals, theta1_vals, J_vals.T, levels=np.logspace(-2, 3, 20))
    ax.set_xlabel('θ₀')
    ax.set_ylabel('θ₁')
    ax.set_title('Cost Function Contours')
    
    plt.tight_layout()
    plt.show()
```

# Main execution

```python
if __name__ == "__main__":
    # Test warmUpExercise
    print("Running warmUpExercise...")
    print("5x5 Identity Matrix:")
    print(warmUpExercise())
    
    # Load data
    data = np.loadtxt('../../notebooks/linear-regression/data/ex1data1.txt', delimiter=',')
    X = data[:, 0]  # Population size in 10,000s
    y = data[:, 1]  # Profit in $10,000s
    m = len(y)  # Number of training examples
    
    # Plot data
    plotData(X, y)
    
    # Add a column of ones to X
    X = np.column_stack((np.ones(m), X))
    
    # Initialize theta
    theta = np.zeros(2)
    
    # Test the cost function
    J = computeCost(X, y, theta)
    print(f'With theta = [0 ; 0]\\nCost computed = {J:.2f}')
    print('Expected cost value (approx) 32.07\\n')
    
    J = computeCost(X, y, np.array([-1, 2]))
    print(f'With theta = [-1 ; 2]\\nCost computed = {J:.2f}')
    print('Expected cost value (approx) 54.24\\n')
    
    # Run gradient descent
    alpha = 0.01
    iterations = 1500
    theta, J_history = gradientDescent(X, y, theta, alpha, iterations)
    
    # Print theta
    print('Theta found by gradient descent:')
    print(f'{theta[0]:.4f}')
    print(f'{theta[1]:.4f}')
    print('Expected theta values (approx)')
    print('-3.6303')
    print('1.1664\\n')
    
    # Plot the linear fit
    plotLinearFit(X, y, theta)
    
    # Plot the cost history
    plotCostHistory(J_history, iterations)
    
    # Plot the cost surface
    plotCostSurface(X, y)
    
    # Predict values for population sizes of 35,000 and 70,000
    predict1 = np.array([1, 3.5]).dot(theta)
    print(f'For population = 35,000, we predict a profit of {predict1*10000:.2f}')
    predict2 = np.array([1, 7.0]).dot(theta)
    print(f'For population = 70,000, we predict a profit of {predict2*10000:.2f}') 
```
