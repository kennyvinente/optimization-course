### Finding the Optimal Step Size for Gradient Descent

### Symbolic Calculation of Optimal Step Size

We want to find the optimal step size $\alpha$ for a single gradient descent step from our starting point $\mathbf{x}_0 = (80, 20)$. We use symbolic mathematics to derive an exact solution.

First, we define our objective function:
$$f(\mathbf{x}) = (x - 47)^2 + (y - 0.1)^2 + 2$$

The gradient of this function is:
$$\nabla f(\mathbf{x}) = \begin{pmatrix} 2(x - 47) \\ 2(y - 0.1) \end{pmatrix}$$

For our starting point $\mathbf{x}_0 = (80, 20)$, the gradient is:
$$\nabla f(\mathbf{x}_0) = \begin{pmatrix} 2(80 - 47) \\ 2(20 - 0.1) \end{pmatrix} = \begin{pmatrix} 66 \\ 39.8 \end{pmatrix}$$

The next point in gradient descent is given by:
$$\mathbf{x}_1 = \mathbf{x}_0 - \alpha \nabla f(\mathbf{x}_0)$$

Substituting our values:
$$\mathbf{x}_1 = \begin{pmatrix} 80 \\ 20 \end{pmatrix} - \alpha \begin{pmatrix} 66 \\ 39.8 \end{pmatrix} = \begin{pmatrix} 80 - 66\alpha \\ 20 - 39.8\alpha \end{pmatrix}$$

Now we define a function $\phi(\alpha)$ which is our objective function evaluated at $\mathbf{x}_1$:
$$\phi(\alpha) = f(\mathbf{x}_1) = (80 - 66\alpha - 47)^2 + (20 - 39.8\alpha - 0.1)^2 + 2$$

Simplifying:
$$\phi(\alpha) = (33 - 66\alpha)^2 + (19.9 - 39.8\alpha)^2 + 2$$

To find the optimal $\alpha$, we take the derivative of $\phi(\alpha)$ with respect to $\alpha$ and set it to zero:
$$\frac{d\phi}{d\alpha} = -2 \cdot 66 \cdot (33 - 66\alpha) - 2 \cdot 39.8 \cdot (19.9 - 39.8\alpha) = 0$$

Simplifying:
$$-132(33 - 66\alpha) - 79.6(19.9 - 39.8\alpha) = 0$$
$$-4356 + 8712\alpha - 1584.04 + 3168.08\alpha = 0$$
$$11880.08\alpha = 5940.04$$
$$\alpha = 0.5$$

Therefore, the optimal step size for the first iteration of gradient descent is 0.5.

```python
def gradient_descent(f, grad_f, x0, alpha_type='fixed', alpha=0.01, 
                     max_steps=1000, tolerance=1e-6, 
                     c1=1e-4, c2=0.9, rho=0.5, max_line_search_iter=20):
    """
    Algoritmo de gradiente descendente com opções para taxa de aprendizado fixa 
    e busca linear (condições de Armijo e Wolfe).
    
    Parâmetros:
    f: função objetivo
    grad_f: gradiente da função
    x0: ponto inicial
    alpha_type: 'fixed' (taxa fixa) ou 'backtracking' (busca linear)
    alpha: taxa de aprendizado inicial
    max_steps: máximo de iterações
    tolerance: tolerância para convergência
    c1, c2: parâmetros para condições de Wolfe (0 < c1 < c2 < 1)
    rho: fator de redução na busca linear
    max_line_search_iter: máximo de iterações na busca linear
    
    Retorna:
    x: ponto ótimo
    f_values: valores da função durante as iterações
    path: caminho percorrido
    """
    x = np.array(x0, dtype=float)
    f_values = [f(x)]
    path = [x.copy()]
    
    for _ in range(max_steps):
        grad = grad_f(x)
        if np.linalg.norm(grad) < tolerance:
            break
            
        if alpha_type == 'fixed':
            alpha_k = alpha
        elif alpha_type == 'backtracking':
            alpha_k = alpha
            for _ in range(max_line_search_iter):
                x_new = x - alpha_k * grad
                # Condição de Armijo
                if f(x_new) <= f(x) + c1 * alpha_k * np.dot(grad, -grad):
                    # Condição de Wolfe (opcional)
                    if np.dot(grad_f(x_new), -grad) >= c2 * np.dot(grad, -grad):
                        break
                alpha_k *= rho
                
        x = x - alpha_k * grad
        f_values.append(f(x))
        path.append(x.copy())
        
    return x, np.array(f_values), np.array(path)

# Funções auxiliares para visualização
def plot_contour(f, path, title, x_range, y_range):
    x = np.linspace(x_range[0], x_range[1], 100)
    y = np.linspace(y_range[0], y_range[1], 100)
    X, Y = np.meshgrid(x, y)
    Z = np.zeros_like(X)
    
    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            Z[i, j] = f(np.array([X[i, j], Y[i, j]]))
    
    plt.figure(figsize=(10, 6))
    plt.contourf(X, Y, Z, 50, cmap='viridis')
    plt.colorbar()
    plt.plot(path[:, 0], path[:, 1], 'r.-', linewidth=1, markersize=3)
    plt.scatter(path[0, 0], path[0, 1], c='blue', s=50, label='Início')
    plt.scatter(path[-1, 0], path[-1, 1], c='red', s=50, label='Fim')
    plt.title(title)
    plt.legend()
    plt.show()

def plot_3d(f, path, title, x_range, y_range):
    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')
    
    x = np.linspace(x_range[0], x_range[1], 100)
    y = np.linspace(y_range[0], y_range[1], 100)
    X, Y = np.meshgrid(x, y)
    
    # Correção: cálculo eficiente do Z
    Z = np.zeros(X.shape)
    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            Z[i, j] = f([X[i, j], Y[i, j]])
    
    ax.plot_surface(X, Y, Z, cmap=cm.viridis, alpha=0.6)
    
    # Correção: cálculo dos valores Z para o caminho
    path_z = np.array([f(p) for p in path])
    ax.plot(path[:, 0], path[:, 1], path_z, 'r.-', linewidth=1)
    ax.scatter(path[0, 0], path[0, 1], path_z[0], c='blue', s=50, label='Início')
    ax.scatter(path[-1, 0], path[-1, 1], path_z[-1], c='red', s=50, label='Fim')
    
    ax.set_title(title)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('f(X,Y)')
    plt.legend()
    plt.show()

def plot_convergence(f_values, title):
    plt.figure(figsize=(10, 6))
    plt.plot(f_values, 'b-', linewidth=1.5)
    plt.yscale('log')
    plt.title(title)
    plt.xlabel('Iteração')
    plt.ylabel('Valor da Função (log)')
    plt.grid(True)
    plt.show()
```

```python
# 1. Função Rosenbrock (n=5)
def rosenbrock(x):
    return sum(100*(x[i+1] - x[i]**2)**2 + (1 - x[i])**2 for i in range(len(x)-1))

def grad_rosenbrock(x):
    n = len(x)
    grad = np.zeros(n)
    for i in range(n-1):
        grad[i] += -400 * x[i] * (x[i+1] - x[i]**2) - 2 * (1 - x[i])
        grad[i+1] += 200 * (x[i+1] - x[i]**2)
    return grad

# 2. Função Booth
def booth(x):
    return (x[0] + 2*x[1] - 7)**2 + (2*x[0] + x[1] - 5)**2

def grad_booth(x):
    df_dx = 2*(x[0] + 2*x[1] - 7) + 4*(2*x[0] + x[1] - 5)
    df_dy = 4*(x[0] + 2*x[1] - 7) + 2*(2*x[0] + x[1] - 5)
    return np.array([df_dx, df_dy])

# 3. Função Himmelblau
def himmelblau(x):
    return (x[0]**2 + x[1] - 11)**2 + (x[0] + x[1]**2 - 7)**2

def grad_himmelblau(x):
    df_dx = 4*x[0]*(x[0]**2 + x[1] - 11) + 2*(x[0] + x[1]**2 - 7)
    df_dy = 2*(x[0]**2 + x[1] - 11) + 4*x[1]*(x[0] + x[1]**2 - 7)
    return np.array([df_dx, df_dy])

# 4. Função Three-Hump Camel
def camel(x):
    return 2*x[0]**2 - 1.05*x[0]**4 + x[0]**6/6 + x[0]*x[1] + x[1]**2

def grad_camel(x):
    df_dx = 4*x[0] - 4.2*x[0]**3 + x[0]**5 + x[1]
    df_dy = x[0] + 2*x[1]
    return np.array([df_dx, df_dy])

# 5. Função McCormick
def mccormick(x):
    return np.sin(x[0] + x[1]) + (x[0] - x[1])**2 - 1.5*x[0] + 2.5*x[1] + 1

def grad_mccormick(x):
    df_dx = np.cos(x[0] + x[1]) + 2*(x[0] - x[1]) - 1.5
    df_dy = np.cos(x[0] + x[1]) - 2*(x[0] - x[1]) + 2.5
    return np.array([df_dx, df_dy])
```

```python
# Configurações comuns
max_steps = 1000
tolerance = 1e-6
backtracking_params = {'alpha_type': 'backtracking', 'alpha': 1.0, 'c1': 1e-4, 'c2': 0.9, 'rho': 0.5}

# 1. Rosenbrock (n=5)
x0_rosen = [-1.5, -1.0, 0.5, 1.5, 2.0]
_, fvals, path = gradient_descent(rosenbrock, grad_rosenbrock, x0_rosen, 
                                 alpha=0.0001, max_steps=max_steps, tolerance=tolerance)
plot_convergence(fvals, "Rosenbrock - Convergência (α fixo=0.0001)")

# 2. Booth
## Taxa fixa com diferentes alphas
for alpha in [0.01, 0.1]:
    _, fvals, path = gradient_descent(booth, grad_booth, [5, 5], 
                                     alpha=alpha, max_steps=max_steps, tolerance=tolerance)
    plot_contour(lambda x: booth(x), path, f"Booth - Trajetória (α fixo={alpha})", [-10, 10], [-10, 10])
    plot_convergence(fvals, f"Booth - Convergência (α fixo={alpha})")

## Busca linear
_, fvals, path = gradient_descent(booth, grad_booth, [5, 5], max_steps=max_steps, 
                                  tolerance=tolerance, **backtracking_params)
plot_contour(lambda x: booth(x), path, "Booth - Trajetória (Busca Linear)", [-10, 10], [-10, 10])
plot_convergence(fvals, "Booth - Convergência (Busca Linear)")

# 3. Himmelblau
## Pontos iniciais diferentes
for x0 in [[0, 0], [4, -2]]:
    _, fvals, path = gradient_descent(himmelblau, grad_himmelblau, x0, 
                                     alpha=0.01, max_steps=max_steps, tolerance=tolerance)
    #plot_3d(lambda x: himmelblau(x.T), path, f"Himmelblau - Superfície (Início {x0})", [-5, 5], [-5, 5])
    # Chamada correta:
    plot_3d(himmelblau, path, f"Himmelblau - Superfície (Início {x0})", [-5, 5], [-5, 5])
    plot_convergence(fvals, f"Himmelblau - Convergência (Início {x0})")

# 4. Three-Hump Camel
## Taxa fixa vs busca linear
_, fvals_fixed, path_fixed = gradient_descent(camel, grad_camel, [1, 1], 
                                            alpha=0.05, max_steps=max_steps, tolerance=tolerance)
_, fvals_bt, path_bt = gradient_descent(camel, grad_camel, [1, 1], max_steps=max_steps, 
                                      tolerance=tolerance, **backtracking_params)

plt.figure(figsize=(10, 6))
plt.semilogy(fvals_fixed, 'b-', label='α fixo=0.05')
plt.semilogy(fvals_bt, 'r-', label='Busca Linear')
plt.title("Three-Hump Camel - Comparação de Métodos")
plt.xlabel('Iteração')
plt.ylabel('Valor da Função (log)')
plt.legend()
plt.grid(True)
plt.show()

# 5. McCormick
## Busca linear com condições de Wolfe
_, fvals, path = gradient_descent(mccormick, grad_mccormick, [0, 0], max_steps=max_steps, 
                                  tolerance=tolerance, **backtracking_params)
plot_3d(mccormick, path, "McCormick - Trajetória 3D", [-3, 3], [-3, 3])
plot_convergence(fvals, "McCormick - Convergência (Busca Linear)")
```
